```
#yeni bir txt dosyasi oluturuluyor
dosya1 = open('metin_dosyasi.txt', 'w')
dosya1.write("Deneme satiri\nHElloWorld")
dosya1.close()


def function1():
    print("dosya adi giriniz 'ornek.txt': ")
    dosyaAdi = input()
    #txt dosyasini oku
    arr = ""
    dosya2 = open(dosyaAdi, 'r')  
    for satir in dosya2:  
        arr += satir  
    dosya2.close()  

    #sifrelemek icin alfabeyi tutuyoruz
    alfabe=['a','b' ,'c' , 'd', 'e', 'f', 'g' ,'h' ,'i' ,'j','k' ,'l' ,'m' ,'n' ,'o' ,'p' ,'r' ,'s' ,'t' ,'u' ,'v', 'y' ,'z']
    #anahtar girilecek
    print("Lütfen anahtar giriniz: ")
    anahtar = int(input())
    arr2 = ""
    for i in arr:
        if(i in alfabe):
            arr2 += alfabe[(alfabe.index(i)+anahtar) % len(alfabe)]
        else:
            arr2 += i

    dosya3 = open('sifre.txt', 'w')
    dosya3.write(arr2)
    dosya3.close()
    
def function2():
    print("dosya adi giriniz 'ornek.txt': ")
    dosyaAdi = input()
    #txt dosyasini oku
    arr = ""
    dosya2 = open(dosyaAdi, 'r')  
    for satir in dosya2:  
        arr += satir  
    dosya2.close() 
    #sifrelemek icin alfabeyi tutuyoruz
    alfabe=['a','b' ,'c' , 'd', 'e', 'f', 'g' ,'h' ,'i' ,'j','k' ,'l' ,'m' ,'n' ,'o' ,'p' ,'r' ,'s' ,'t' ,'u' ,'v', 'y' ,'z']
    #anahtar girilecek
    print("Lütfen anahtar giriniz: ")
    anahtar = int(input())
    arr2 = ""
    for i in arr:
        if(i in alfabe):
            arr2 += alfabe[(alfabe.index(i)-anahtar) % len(alfabe)]
        else:
            arr2 += i
    print("Cozulmus Sifre:\n"+arr2)

def function3():
    print("dosya adi giriniz 'ornek.txt': ")
    dosyaAdi = input()
    #txt dosyasini oku
    arr = ""
    dosya2 = open(dosyaAdi, 'r')  
    for satir in dosya2:  
        arr += satir  
    dosya2.close() 
    #sifrelemek icin alfabeyi tutuyoruz
    alfabe=['a','b' ,'c' , 'd', 'e', 'f', 'g' ,'h' ,'i' ,'j','k' ,'l' ,'m' ,'n' ,'o' ,'p' ,'r' ,'s' ,'t' ,'u' ,'v', 'y' ,'z']
    #anahtar girilecek
    for anahtar in range(len(alfabe)):
        arr2 = ""
        for i in arr:
            if(i in alfabe):
                arr2 += alfabe[(alfabe.index(i)-anahtar) % len(alfabe)]
            else:
                arr2 += i
        print("Anahtar " + str(anahtar) + " icin cozulmus sifre:\n" + arr2)    
        
secim = 0
while(secim != 4):
    print("\n1-Dosya Sifrele\n2-Sifre coz\n3-Anahtarsız Sifre coz\n4-cikis\n")
    secim = int(input())
    if(secim == 1):
        function1()
    if(secim == 2):
        function2()
    if(secim == 3):
        function3()
```
